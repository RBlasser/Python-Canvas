```python
import pandas as pd
import numpy as np
import pandas_datareader
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
PG = wb.DataReader('PG', data_source = 'iex', start = '2015-01-01')
```

```python
PG.head()
```

```python
PG.tail()
```

```python
PG['simple_return'] = (PG['close'] / PG['close'].shift(1))-1
```

```python
PG[['close','simple_return']]
```

```python
PG_plot = PG[['close','simple_return']]
```

```python
PG_plot.plot(figsize=(10,4), linewidth=3)
plt.show()
```

```python
PG_plot2 = PG[['simple_return']]
```

```python
PG_plot2.plot(figsize=(8,5))
plt.show()
```

```python
PG_plot3 = PG[['close']]
PG_plot3.plot(figsize=(8,5))
plt.show()
```

```python
PG_avg = PG['simple_return'].mean()
```

```python
PG_avg
```

```python
# 250 is the number of active days, excluding holidays, weekends, etc.
PG_avg_annual = PG['simple_return'].mean() * 250
```

```python
PG_avg_annual
```

```python
print (str(round(PG_avg_annual,4) * 100) + '%')
```

```python
PG
```

```python
PG['log_return'] = np.log(PG['close']/PG['close'].shift(1))
```

```python
PG['log_return']
```

```python
PG
```

```python
print (PG['log_return'])
```

```python
PG_plot4 = PG[['simple_return','log_return']]
PG_plot4.plot(figsize=(8,3))
plt.show()
```

```python
# 250 is the number of active days, excluding holidays, weekends, etc.
PG_avg_annual = PG['simple_return'].mean() * 250

# 250 is the number of active days, excluding holidays, weekends, etc.
PG_avg_annualog = PG['log_return'].mean() * 250
```

```python
print (str("PG annual simple return: " + str(round(PG_avg_annual,4) * 100) + '%'))
print (str("PG annual log return: " + str(round(PG_avg_annualog,4) * 100) + '%'))
```
