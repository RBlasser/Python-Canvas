# Portfolio Risk

```python
weights = np.array([0.50,0.50])
```

```python
weights
```

```python
# Portfolio variance
## This formula works for building a cov matrix, resulting in one number
### weights.T means that it will transpose the vector to make this calculation
```

```python
pfolio_var = np.dot(weights.T, np.dot(sec_returns_log.cov()*250, weights))
pfolio_var
```

```python
# Portfolio volatility
```

```python
pfolio_vol = np.dot(weights.T, np.dot(sec_returns_log.cov()*250, weights)) **.5
pfolio_vol
```

```python
print (str(round(pfolio_vol,5)*100) + '%')
```

```python
weights
```

# Diversifiable (non systematic) & Un-diversifiable (systematic) Risks

```python
weights[0]
```

```python
weights[1]
```

```python
PG_var_annual
```

```python
BEI_var_annual
```

```python
# Diversifiable Risk "unsystematic risk" = portfolio variance - (difference among weighted annual variance of each security)
## The weighted annual variance per security is the result of powering to 2 the annual variance
```

```python
div_risk = pfolio_var - (weights[0]**2 * PG_var_annual) - (weights[1]**2 * BEI_var_annual)
```

```python
div_risk
```

```python
print (str(round(div_risk,5)*100) + '%')
```

```python
# Un Diversifiable Risk "systematic risk" = Σ weighted annual variance
```

```python
nondiv_risk1 = pfolio_var - div_risk
```

```python
nondiv_risk1
```

```python
print (str(round(nondiv_risk1,5)*100)+'%')
```

```python
nondiv_risk2 = weights[0]**2 * PG_var_annual + weights[1]**2 * BEI_var_annual
```

```python
nondiv_risk2
```

```python
print (str(round(nondiv_risk2,5)*100)+'%')
```

```python
# Validation (boolean) if both calculations result in the same number
```

```python
nondiv_risk1 == nondiv_risk2
```

# Beta

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb

tickers = ['PG','^GSPC']
data = pd.DataFrame()
for t in tickers:
    data[t]= wb.DataReader(t, data_source = 'yahoo', start = '2012-1-1', end = '2016-12-31')['Adj Close']
```

```python
data.head()
```

```python
data.tail()
```

```python
sec_returns = np.log(data/data.shift(1))
```

```python
sec_returns.head()
```

```python
sec_returns.tail()
```

```python
#Returns a covariance matrix for PG and GSPC
cov = sec_returns.cov()*250
cov
```

```python
#Returns by coordinates (0,1) the covariance between PG and GSPC
cov_with_market = cov.iloc[0,1]
cov_with_market
```

```python
#Returns the annual variance for the GSPC
market_var = sec_returns['^GSPC'].var() *250
market_var
```

```python
#Returns the volatility (std dev)
market_stdv = np.sqrt(market_var)
market_stdv
```

# PG Beta

```python
PG_beta = cov_with_market/market_var
PG_beta
```

# Calculate de expected return for PG, based on CAPM


```python
# The expected return (Ri) is equal to the Risk-free asset (Rf) + Beta of the security times the Risk Premium (Rm - Rf)
# The best security, closer to be risk-free, is the yield for the 10 Year US Treasury Bond (currently 2.5%)
# Research has demonstrated that the Risk Premium is usually between 4.5% and 5.5%, so 5% is somewhere in the middle
```

```python
PG_er = 0.025 + PG_beta*0.05
PG_er
```

```python
print(round(PG_er*100,2), '%')
```

```python
sec_returns['PG'].mean()
```

# Sharpe Ratio 

```python
# Sharpe Ratio formula
# Sharpe Ratio = (Return on Security - Risk Free Rate)/ Standar Deviation of the Security
```

```python
Sharpe = (PG_er - 0.025) / (sec_returns['PG'].std() * 250 **0.5)
Sharpe
```
